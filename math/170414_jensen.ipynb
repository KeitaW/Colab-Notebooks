# [イェンゼンの不等式の三通りの証明](http://mathtrain.jp/jensen_proof)より

イェンゼンの不等式（Jensen，凸関数の不等式）$f(x)$が凸関数のとき，任意に$\lambda_i \geq 0$, $x_i, (i=1,\dots, n), \sum_{i=1}^n \lambda_i = 1$に対して
$$
\sum_{i=1}^n\lambda_i f(x_i) \geq f(\sum_{i=1}^n\lambda_ix_i)
$$

特に２変数の場合
$$
\lambda_1 f(x_1) + \lambda_2 f(x_2) \geq f(\lambda_1 x_1 + \lambda_2 x_2)
$$

を詳しく考察してみる

$f(x)$が凸関数である．

$\equiv$

* ${}^\forall x_1, x_2, \lambda \ (0 \geq \lambda \geq 1)$に対して 
$$
\lambda f(x_1) + (1-\lambda)f(x_2) \geq f(\lambda x_1 + (1-\lambda) x_2)
$$


or 
* ${}^\forall x_1, x_2$に対して２点$(x_1, f(x_1)), (x_2, f(x_2))$を結ぶ線分が関数の上側にある．

or
* 二階微分$f''(x)$が存在して$0$以上である


```julia
using PyPlot
```

```julia
t = -10:10; f(t) = t.^2;
x1 = -7.5; y1 = f(x1); x2 = 5; y2 = f(x2)
λ1 = 0.5; λ2 = 0.5; x3 = λ1*x1 + λ2*x2
fig, ax = subplots(1, 1)
ax[:set_ylim](0, 80)
ax[:plot](t, f(t), label="LHS"); ax[:plot]([x1, x2], [y1, y2])
ax[:scatter](x3, f(x3)); ax[:annotate](L"$f(\lambda_1x_1+\lambda_2x_2)$", xy=(x3, f(x3)+2.5), fontsize=20)
ax[:scatter](x3, λ1*f(x1)+λ2*f(x2)); ax[:annotate](L"$\lambda_1f(x_1)+\lambda_2f(x_2)$", xy=(x3, λ1*f(x1)+λ2*f(x2)+1.5), fontsize=20)
ax[:vlines](x1, 0, y1, linestyles="dashed"); ax[:vlines](x2, 0, y2, linestyles="dashed")
ax[:set_xticks]([x1, x2]); ax[:set_xticklabels](["x1", "x2"])
ax[:annotate](L"$f(t)$", xy=(8, 50), fontsize=20)
```

* $n=2$のイェンセンの不等式は凸関数の性質から明らか
* $n=k$のときのイェンゼンの不等が成立すると仮定すると$n=k+1$のときにも不等式が成立することを示せる

$$
\sum_{i=1}^{k+1}\lambda_i f(x_i) = \sum_{i=1}^{k+1} \lambda_if(x_i) + \lambda_{k+1}f(x_{k+1}) \\
= A \sum_{i=1}^{k}\frac{\lambda_i}{A}f(x_i) + \lambda_{k+1}f(x_{k+1}) \quad A \equiv \sum_{i=1}^k \lambda_i \\
\geq  A f\left( \sum_{i=1^k}\frac{\lambda_ix_i}{A}\right) + \lambda_{k+1}f(x_{k+1}) 
$$
